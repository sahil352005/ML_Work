```python
##  Create a perceptron model to learn the AND gate using randomly initialized weights and bias.
##  Track the weight updates for each epoch using NumPy.
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Input and Output for AND Gate
X = np.array([
    [0, 0],
    [0, 1],
    [1, 0],
    [1, 1]
])
y = np.array([0, 0, 0, 1])

# Initialize weights and bias randomly
np.random.seed(42)  # reproducibility
weights = np.random.randn(2)
bias = np.random.randn()
learning_rate = 0.1
epochs = 10

# Step function as activation
def step_function(x):
    return 1 if x >= 0 else 0

# To store misclassifications per epoch
misclassifications = []

print("Initial Weights:", weights)
print("Initial Bias:", bias)
print("-" * 40)

# Training the perceptron
for epoch in range(epochs):
    errors = 0
    print(f"Epoch {epoch + 1}")
    for i in range(len(X)):
        linear_output = np.dot(X[i], weights) + bias
        prediction = step_function(linear_output)
        error = y[i] - prediction

        # Update weights and bias
        weights += learning_rate * error * X[i]
        bias += learning_rate * error

        if error != 0:
            errors += 1

        print(f"  Input: {X[i]} | Target: {y[i]} | Predicted: {prediction} | Error: {error}")
        print(f"  Updated Weights: {weights} | Updated Bias: {bias}")
    misclassifications.append(errors)
    print(f"Misclassifications in Epoch {epoch+1}: {errors}")
    print("-" * 40)

# Final weights and bias
print("Final Weights:", weights)
print("Final Bias:", bias)

# Plotting misclassifications over epochs
plt.plot(range(1, epochs + 1), misclassifications, marker='o', linestyle='-', color='blue')
plt.title('Misclassifications per Epoch (AND Gate with Random Init)')
plt.xlabel('Epoch')
plt.ylabel('Number of Misclassifications')
plt.grid(True)
plt.tight_layout()
plt.show()
```
